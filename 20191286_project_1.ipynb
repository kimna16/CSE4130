```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

def drawLine(w, col):
    x = np.arange(-5,10.1,0.1,dtype=None)
    #y = -w(1)/w(3)-w(2)/w(3)*x
    y=-w[0][0]/w[0][2]-w[0][1]/w[0][2]*x
    plt.plot(x, y, col)    
      
    
def drawDataHyperplane(x_c1, x_c2, w, col):
    #dataDraw(x_c1, x_c2)
    
    plt.plot(x_c1[:,1], x_c1[:,2], '+b')
    plt.plot(x_c2[:,1], x_c2[:,2], 'or') 
    drawLine(w, col)
    plt.plot()
    #plt.show()

def perceptron_mis_yx(y, x, w, n_epoch, eta):
    n_sample=len(y)//2
    x_c1=x[1:n_sample,:]
    x_c2=x[n_sample+1:n_sample*2,:]
    #yx=np.tile(y, (1, 3))*np.transpose(x)
    yx=np.multiply(np.tile(y, (1, 3)),x)
    
    print("----- Perceptron_mis_yx() started -----\n")
    for e in range(1,n_epoch+1):
        #wyx=dot(np.tile(w,(n_sample*2,1)),yx,2)
        wyx=np.transpose(np.diag(np.dot(np.tile(w,(n_sample*2,1)),np.transpose(yx))))
        
        yx_misclassified=np.array(yx[wyx<0,:])
        
        w=w+sum(yx_misclassified)
        
        err=0
        err=sum(wyx<0)
        #err=0
        #for a in wyx:
        #    if (a<0):
        #        err+=1
        
        print("err = {}\n".format(err))
        #plt.figure(2)
        #if (e==1):
        plt.xlim([-5, 10])
        plt.ylim([-5, 10])
        #plt.axis('equal')
        drawDataHyperplane(x_c1, x_c2, w, 'k')
        plt.plot()
        #plt.show()
        if (err == 0):
            break
    plt.show()
    return e



def perceptron_l2(y, x, w, n_epoch, eta):
    n_sample=len(y)//2
    x_c1=x[1:n_sample,:]
    x_c2=x[n_sample+1:n_sample*2,:]
    
    print("----- Perceptron_l2() started -----\n")
    for e in range(1,n_epoch+1):
        # wx = dot(repmat(w,n_sample*2,1),x,2)
        wx=np.transpose(np.diag(np.dot(np.tile(w,(n_sample*2,1)),np.transpose(x))))
        
        # diff=repmat((y-wx),1,3).*x
        #print(x)
        diff=np.multiply(np.tile(y-wx[:,None], (1,3)),x)
        #print(diff)
        
        #print(eta*np.sum(diff))
        #print(w)
        # w = w + eta*sum(diff)
        w=w+eta*sum(diff)
        #print(w)
 
        # err1 = 1/2*sum((y - wx).^2)
        err1=np.sum(np.power(y-wx[:,None],2))*1/2
        
        # err2 = sum(y.*wx<0)
        z=np.multiply(y,wx[:,None])
        err2=np.sum(z<0)
    
        print("err({:d}) = {:f} {:d}\n".format(e,err1,err2))
        #plt.figure(3)
        plt.xlim([-5,10])
        plt.ylim([-5,10])
        #plt.axis([-5,10,-5,10])
        #plt.axis('equal')
        drawDataHyperplane(x_c1, x_c2, w, 'k')
        plt.plot()
        if (err2 == 0):
            break
    plt.show()
    return e


def SVM(y, x, w, n_epoch, eta):
    yx=np.multiply(np.tile(y,(1,3)),x)
    print("svm")
    for e in range(1,n_epoch+1):
        #wyx = dot(repmat(w,n_sample*2,1),yx,2);
        wyx=np.transpose(np.diag(np.dot(np.tile(w,(n_sample*2,1)),np.transpose(yx))))
        
        # err_idx = wyx<0;
        err_idx = wyx<1
        
        lamda = 0.01
        # w = w + eta*(-2*lamda*w + sum(yx(err_idx,:)));
        w = w + eta*(-2*lamda*w*(n_sample*2) + sum(yx[err_idx,:]))
        # w = w + eta*(sum(yx(err_idx,:)));
        
        err1 = np.sum(err_idx)
        err2 = np.sum(wyx[err_idx-1])
        a=math.sqrt(np.sum(np.power(w,2)))
        b=math.sqrt(sum(np.power(w.flatten()[1:3],2)))
        print("err({:d}) = {:f} {:f}, w= {:f} {:f} {:f}, |w|={:f}, |w23|={:f}\n".format(e, err1, err2, w[0][0],w[0][1],w[0][2],a,b))
        #plt.figure(5)
        plt.xlim([-5,10])
        plt.ylim([-5,10]) 
        #plt.axis('equal')
        drawDataHyperplane(x_c1, x_c2, w, 'k')
        plt.plot()
    plt.show()
    return


n_sample = 20;
# x_c1 = mvnrnd([0, 0], [1, .95; .95, 1.], n_sample);
# x_c2 = mvnrnd([1, 5], [1, .85; .85, 1.], n_sample);
# dlmwrite('x_c1', x_c1);
# dlmwrite('x_c2', x_c2);
x_c1 = pd.read_csv('x_c1.csv',delimiter=',',header=None)
x_c2 = pd.read_csv('x_c2.csv',delimiter=',',header=None)

#np.ones? np.ones_like?

# add bias term
#np.array를 해야 하나?
x_c1 = np.hstack((np.ones((n_sample,1)),x_c1)); 
x_c2 = np.hstack((np.ones((n_sample,1)),x_c2)); 

# label
y1 = np.ones((n_sample,1))*-1
y2 = np.ones((n_sample,1))

x=np.vstack((x_c1,x_c2));
y=np.vstack((y1,y2));

w = np.random.rand(1,3);

#plt.figure(1) 
#plt.xlim([-5, 10])
#plt.ylim([-5, 10])
#plt.axis('equal')
#drawDataHyperplane(x_c1, x_c2, w, 'k')

# w <- w+sum(misclassified yx), Fig(2)
n_epoch = 100
eta = 0.001
ret1 = perceptron_mis_yx(y, x, w, n_epoch, eta)
 
# w <- w+sum(squared error), Fig(3)
n_epoch = 100
eta = 0.001
ret2 = perceptron_l2(y, x, w, n_epoch, eta)


n_epoch = 300
eta = 0.001
SVM(y, x, w, n_epoch, eta)

#print("ret1 = %d, ret2 = %d\n".format(ret1, ret2))
```
